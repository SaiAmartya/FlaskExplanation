# What is Flask?

*   Flask is a lightweight web framework in python, that allows developers to creat API's and applications without much setup. This can let you easily use python for back end complex logic. It is commonly used to build REST Api's, which enables communication between client and server through HTTP methods like GET, POST, PUT and DELETE.

**Why use Flask instead of Javascript?**

**Flask:**

*   Server-side development
*   Process user input
*   Manage databases
*   Implement API’s to interact with client
*   Handle security/authentication
*   (You can now use python for the back-end, ideal for complex backend logic!)

**Javascript:**

*   Client-side interactivity
*   Render dynamic content
*   Interact with the server via API’s

###Now, you might have ran into the 405 **POST** error. This is because we didn't state that we might use the **POST** method.

**But, what are HTTP methods?**

Common **HTTP** methods include;
- **GET**
  - Used to **retrieve** data from server
  - Doesn't modify data
  - Often used to load webpages or fetching lists of items
- **POST**
  - Used to **send** data to server
  - Can update/modify resources
  - Often used to submit forms and upload files
- **PUT** (We won't talk about this one)
  - Used to **update** or **create** a resource
  - Often used for full resource updates
- **PATCH** (We wont really talk about this one that much)
  - Can partially update a resource

In the bellow code, we now have mentioned that we will also use the **POST** method.

*Note: By default, the only method will be GET, but if we are going to use any more, then we must mention all of those that will be used.*

*app.py*
```
from flask import Flask, request, render_template, redirect, url_for
# imports flask, request - for user input information, render_template for html files, and redirect along with url_for to send the user to another webpage
app = Flask(__name__)
# creates an application with flask

@app.route("/login", methods=["POST", "GET"]) #creates a webpage with its own URL and has specified methods that can be used on the page to update information
def login():
    if request.method == "POST": #if the requested information is recieved from the user instead of the server (user presses submit)
        first_name = request.form["fnm"] #requests for the information the user filled in for their first name
        last_name = request.form["lnm"] #requests for the information the user filled in for their last name
        return redirect(url_for("user", first_name=first_name, last_name=last_name)) #the user is redirected to a specfic URL
    else: #Gets the login page if the user has yet to access the page, (loads the page otherwise)
        return render_template("login.html")
    
@app.route("/user") #new URL, which is the one where the user is redirected to
def user(): #function is associated is connected to the URL the user is redirected to
    first_name = request.args.get("first_name") #gets the first name the user typed in on the login page
    last_name = request.args.get("last_name") #gets the first name the user typed in on the login page
    return f"<h1>Welcome, {first_name} {last_name} </h1>"

if __name__ == "__main__":
    app.run(debug=True)
    # runs flask
```
